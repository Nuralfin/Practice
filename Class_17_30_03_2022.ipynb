```python
df2 = pd.read_excel(r'C:/Users/pyari/Desktop/crop_2.xlsx')
```

```python
x=df2["wheat"]
y=df2["pulses"]
t=x
plt.scatter(x,y, c=t,s=100, cmap="Paired",marker='d',alpha=0.7)
plt.colorbar()
plt.ylabel("Pulses")
plt.xlabel("wheat")
plt.yticks(rotation=45)
plt.title("Scatter plot")
plt.show()
```

```python
data = [df2["rice"],df2["wheat"],df2["pulses"],df2["Maize"]]
```

```python
data
```

```python
box = plt.boxplot(data,notch=False,vert=True,positions=[1,3,5,7],widths=1.5,labels=['Rice','wheat','Pulses','maize'],patch_artist=True)
colors = ['cyan', 'lightblue', 'lightgreen', 'tan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel("Production")
plt.xlabel("Crops")
        
```
